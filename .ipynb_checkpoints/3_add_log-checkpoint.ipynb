# Predicting the electronic dielectric constant using CGCNN

### Summary:
Dielectric constants are important in designing battery electrolyte. Here, we can try to predict them using CGCNN$^1$ with the dielectric_constant dataset$^2$ from Matminer$^3$ 

**Caveats and targets:**

**Methodology:**  
**Dataset:** dielectric_constant from Matminer  
**X_train:** Features extracted via CGCNN; log(1+X) modification  
**Model:** CGCNN (modified, see below)  
**N:** 1056 (63%/27%/10% train/val/test split); stratified split with binned target to ensure similar train/val/test distribution 

**Key Findings:**  
a) Using log(1+X) modification helps reduce variance with the long tails   
b) Due to small dataset, resultant fit is very sensitive to the way we split train/val/test; stratified splitting ensures that train/test are from similar distribution (Biggest impact on results)  
c) It is extremely crucial to ensure that atleast a few datapoints in the tail are captured in the validation dataset, otherwise the training is not optimal 
d) Using structural features via CGCNN results in a better model compared to using only the composition features with random forest models

**Modifications to the CGCNN code**  
a) The training and validation losses are returned to plot the loss functions as a function of epochs to test bias and variance  
b) Since train/val/test splits are predetermined via stratified splitting, the test parts of the code were edited out  
c) The validation dataset was appended to the training dataset and the overall dataset was passed to the code   
d) The predict.py of the code was used to rerun the best model on the train, validation, and test datasets separately after training the model  
e) To obtain reproducible results, the randomness is seeded and determinism is added

**Results (reproducible)**

**Train  |  Validation  |  Test  |**  
MAE   |  1.238  |  1.066  |  1.061  |  
RMSE  |  7.681  |  3.228  |  2.352  |  
R$^2$ |  0.760  |  0.767  |  0.833  |  


**Issues and how they were addressed:**  
a) Outliers skew the data and increase variance; log(1+X) modification  
b) Since there are only few datapoints, the way we split the train/val/test set affects the model starkly (random_state variable); stratified splitting for train/val/test split

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import sys, subprocess, os
import ast
from pathlib import Path
import json
import shutil

from matminer.datasets import load_dataset
from pymatgen.core.structure import Structure
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, KFold, RepeatedKFold, StratifiedKFold
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression

import warnings
warnings.filterwarnings('ignore')

rs = 82 #random_state variable
```

```python
# Load dataset
df = load_dataset('dielectric_constant')
df = df.dropna(how='all')
df.tail() # Display the dataset
```

```python
# Create the directories to keep the structure file and the id-target file

PROC = Path("data/processed")
CIF_DIR = PROC / "cif"
IN_CSV = CIF_DIR / "all_prop.csv"
CIF_DIR.mkdir(parents=True, exist_ok=True)

def main():
    #Convert the tensors into a scalar for training
    df["e_electronic_scalar"] = df["e_electronic"].apply(lambda x: np.trace(np.array(ast.literal_eval(x)))/3 if isinstance(x, str) else np.nan)
    df["e_total_scalar"] = df["e_total"].apply(lambda x: np.trace(np.array(ast.literal_eval(x)))/3 if isinstance(x, str) else np.nan)

    #Set the target
    df["targets"] = df["e_electronic_scalar"]

    #Obtain the structures and send them in cif format to the CIF_DIR
    material_id, targets = [], []
    for row in df.itertuples(index=False):
        s_material_id = getattr(row, "material_id")
        target = getattr(row, "targets")
        structure = getattr(row, "structure")
        cif_file = CIF_DIR / f"{s_material_id}.cif"
        structure.to(filename=str(cif_file))

        targets.append(float(target))
        material_id.append(s_material_id)

    df2 = pd.DataFrame({"material_id": material_id, "targets": targets})
    df2 = df2.dropna(how='all')
    df2.to_csv(IN_CSV, index=False, header=False)

    print(f"Written {len(material_id)} CIFs to {CIF_DIR} and labels to {IN_CSV}")

if __name__=="__main__":
    main()
```

```python
PROC = Path("data/processed")
CIF_DIR = PROC / "cif"
IN_CSV = CIF_DIR / "all_prop.csv"
TRAIN_CSV = CIF_DIR / "train_prop.csv"
TRAIN_VAL_CSV = CIF_DIR / "train_val_prop.csv"
VAL_CSV = CIF_DIR / "val_prop.csv"
ID_CSV = CIF_DIR / "id_prop.csv"
TEST_CSV = CIF_DIR / "test_prop.csv"
CIF_DIR.mkdir(parents=True, exist_ok=True)

#Create a reproducible train/test split
df = pd.read_csv(IN_CSV, header=None, names=["materials_id", "targets"])
X = df["materials_id"]
y = df["targets"]

#Stratified train+val/test split using binned target
y_binned = pd.qcut(y, q=5, duplicates='drop')  # quantile-based bins
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, stratify=y_binned, test_size=0.1, random_state=rs)

#Stratified train/val split using binned target
y_binned = pd.qcut(y_train_val, q=5, duplicates='drop')  # quantile-based bins
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, stratify=y_binned, test_size=0.3, random_state=rs)

#Apply logarithm to y
y_train_log = np.log1p(y_train)
y_val_log = np.log1p(y_val)
y_test_log = np.log1p(y_test)

#Create separate train and validation dataframes
df2t = pd.DataFrame({"X_train": X_train, "y_train_log": y_train_log}) #y_train_log
df2v = pd.DataFrame({"X_train": X_val, "y_train_log": y_val_log}) #y_val_log
df2t.to_csv(TRAIN_CSV, index=False, header=False)
df2v.to_csv(VAL_CSV, index=False, header=False)

#Append the train and validation datasets
df2 = pd.concat([df2t, df2v], ignore_index=True)
df2.to_csv(TRAIN_VAL_CSV, index=False, header=False)

#Prepare the appended dataset as input for CGCNN
df2.to_csv(ID_CSV, index=False, header=False)

#Create separate test dataframe
df3 = pd.DataFrame({"X_test": X_test, "y_test_log": y_test_log}) #y_test_log
df3.to_csv(TEST_CSV, index=False, header=False)
```

```python
shutil.copy(TRAIN_VAL_CSV, ID_CSV) #redundant if all cells run sequentially

#Train the model
cmd = [
    sys.executable, "external/cgcnn/main.py",
    "--seed", str(rs), "--deterministic", "true",
    "--train-ratio","0.7","--val-ratio","0.3","--test-ratio","0.0",
    "-b","128","--epochs","100",
    "data/processed/cif"  # root_dir containing id_prop.csv, atom_init.json, and cif/
]
env = dict(os.environ)
env["CUDA_VISIBLE_DEVICES"] = "0"  # or omit for default
subprocess.run(cmd, check=True, env=env)
```

```python
#Move all the model files to a separate model directory
MODEL_DIR = Path("model")
MODEL_DIR.mkdir(parents=True, exist_ok=True)
shutil.move("model_best.pth.tar", "model/model_best.pth.tar")
shutil.move("checkpoint.pth.tar", "model/checkpoint.pth.tar")
shutil.move("train_val_loss.csv", "model/train_val_loss.csv")

#Copies the train into ID_CSV
shutil.copy(TRAIN_CSV, ID_CSV) 

#Use the best model to predict the target values for the train dataset
cmd = [
    sys.executable, "external/cgcnn/predict.py",
    "--seed", str(rs), "--deterministic", "true",
    "model/model_best.pth.tar",
    "data/processed/cif"  # root_dir containing id_prop.csv, atom_init.json, and cifs
]
env = dict(os.environ)
env["CUDA_VISIBLE_DEVICES"] = "0"  # or omit for default
subprocess.run(cmd, check=True, env=env)

RESULTS_DIR = Path("results")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

#CGCNN outputs the prediction as "test_results.csv"; converting that to train_results.csv in the results directory
shutil.move("test_results.csv", "results/train_results.csv")
```

```python
#Copies the validation dataset into ID_CSV
shutil.copy(VAL_CSV, ID_CSV) #

#Use the best model to predict the target values for the validation dataset
cmd = [
    sys.executable, "external/cgcnn/predict.py",
    "--seed", str(rs), "--deterministic", "true",
    "model/model_best.pth.tar",
    "data/processed/cif"  # root_dir containing id_prop.csv, atom_init.json, and cifs
]
env = dict(os.environ)
env["CUDA_VISIBLE_DEVICES"] = "0"  # or omit for default
subprocess.run(cmd, check=True, env=env)

RESULTS_DIR = Path("results")
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

#CGCNN outputs the prediction as "test_results.csv"; converting that to val_results.csv in the results directory
shutil.move("test_results.csv", "results/val_results.csv")
```

```python
#Copies the test into ID_CSV
shutil.copy(TEST_CSV, ID_CSV)

#Use the best model to predict the target values for the test dataset
cmd = [
    sys.executable, "external/cgcnn/predict.py",
    "--seed", str(rs), "--deterministic", "true",
    "model/model_best.pth.tar",
    "data/processed/cif"  # root_dir containing id_prop.csv, atom_init.json, and cifs
]
env = dict(os.environ)
env["CUDA_VISIBLE_DEVICES"] = "0"  # or omit for default
subprocess.run(cmd, check=True, env=env)

#CGCNN outputs the prediction as "test_results.csv"; converting that to test_results.csv in the results directory
shutil.move("test_results.csv", "results/test_results.csv")
```

```python
# Function to compute RMSE and R2 metrics
def compute_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = root_mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2

#Read the predicted target values for train, validation, and test datsets stored in results folder
df_train = pd.read_csv("results/train_results.csv", header=None, names=["X_train", "y_train_log_true", "y_train_log_pred"])
df_val = pd.read_csv("results/val_results.csv", header=None, names=["X_val", "y_val_log_true", "y_val_log_pred"])
df_test = pd.read_csv("results/test_results.csv", header=None, names=["X_test", "y_test_log_true", "y_test_log_pred"])

#Remove logarithm from y, both predicted and true values
df_train["y_train_true"]=df_train["y_train_log_true"].apply(lambda x: np.expm1(x))
df_train["y_train_pred"]=df_train["y_train_log_pred"].apply(lambda x: np.expm1(x))
df_val["y_val_true"]=df_val["y_val_log_true"].apply(lambda x: np.expm1(x))
df_val["y_val_pred"]=df_val["y_val_log_pred"].apply(lambda x: np.expm1(x))
df_test["y_test_true"]=df_test["y_test_log_true"].apply(lambda x: np.expm1(x))
df_test["y_test_pred"]=df_test["y_test_log_pred"].apply(lambda x: np.expm1(x))

#Compute the metrics for train, validation, and test datasets
train_mae, train_rmse, train_r2 = compute_metrics(df_train["y_train_true"], df_train["y_train_pred"])
val_mae, val_rmse, val_r2 = compute_metrics(df_val["y_val_true"], df_val["y_val_pred"])
test_mae, test_rmse, test_r2 = compute_metrics(df_test["y_test_true"], df_test["y_test_pred"])

#Create DataFrame containing metrics
df_metrics_total = pd.DataFrame({
    'Train': [train_mae, train_rmse, train_r2],
    'Validation': [val_mae, val_rmse, val_r2],
    'Test': [test_mae, test_rmse, test_r2]
}, index=['MAE', 'RMSE', 'R$^2$'])

#Dump metrics into a separate file in results directory
METRICS_CSV = RESULTS_DIR / "metrics.csv"
df_metrics_total.to_csv(METRICS_CSV)

#Display the table
display(df_metrics_total)
```

```python
#Create parity plot in the results directory
PARITY_PDF = RESULTS_DIR / "elec_parity.pdf"

# Parity plot
plt.figure(figsize=(6,6))
plt.scatter(df_train["y_train_true"], df_train["y_train_pred"], alpha=0.6, color='b', marker='o', label='Train')
plt.scatter(df_val["y_val_true"], df_val["y_val_pred"], alpha=0.6, color='g', marker='o', label='Validation')
plt.scatter(df_test["y_test_true"], df_test["y_test_pred"], alpha=0.6, color='r', marker='o', label='Test')
mn = min(min(df_train["y_train_true"]), min(df_train["y_train_pred"]), min(df_test["y_test_true"]), min(df_test["y_test_pred"]), min(df_val["y_val_true"]), min(df_val["y_val_pred"])) 
mx = max(max(df_train["y_train_true"]), max(df_train["y_train_pred"]), max(df_test["y_test_true"]), max(df_test["y_test_pred"]), max(df_val["y_val_true"]), max(df_val["y_val_pred"]))
plt.plot([mn, mx], [mn, mx], '--k')
plt.xlabel("True Electronic Dielectric Constant")
plt.ylabel("Predicted Electronic Dielectric Constant")
plt.title("Parity Plot for Electronic Dielectric Constant")
plt.legend()
plt.tight_layout()
plt.savefig(PARITY_PDF)
plt.show()
```
